# Programma degli incontri

## Prima lezione: 7-9 Marzo 2025 

@ Istituto Lama Tzong Khapa di Pomaia

### Gli studi psicologici e le pratiche buddiste di meditazione e compassione (I parte)

**Dott. Joan Dombon**

- **Obiettivi** Fornire i fondamenti concettuali relativi al funzionamento della mente nella tradizione buddhista. Introdurre alle pratiche contemplative in termini teorico esperienziali.
- **Contenuti** Introduzione ai contenuti del Corso, con focus sul rapporto tra Psicologia e Buddhismo. Presentazione di alcuni principi buddhisti come: le 4 nobili verità, ottuplice sentiero, i tre tipi di sofferenza. Presentazione dei corsisti. Il ruolo della concentrazione e della consapevolezza nella pratica meditativa.

**Ven. Constance Miller**

- **Obiettivi** Sperimentare gli elementi fondamentali della pratica meditativa: Attenzione
sostenuta, consapevolezza, motivazione. 
- **Contenuti**  Componenti principali della meditazione: attenzione sostenuta,  consapevolezza, cautela. L’importanza dell’atteggiamento basilare: gentilezza, accoglienza, umiltà. Focus sul significato di motivazione e le sue caratteristiche. Periodi pratici di meditazione guidata sono alternati con spiegazioni teoriche e discussione.

## Seconda lezione: 14 Marzo 2025 

@ UNIFI

### La tradizione Mahāyāna e la compassione. Implicazionisul piano professionale. 

**Prof. Corrado Caudek**

- **Obiettivi** Far conoscere il reale significato della meditazione nelle pratiche buddiste
Familiarizzare con la ricerca in campo occidentale sulla compassione e la mindfulness. 

- **Contenuti** Incontri tra Culture: Compassione nell’ambito Buddista e Secolare. Sistematizzazioni Mahāyāna. Che cosa viene coltivato? Metodi di coltivazione della compassione. La meditazione. Meditazione Tonglen. La perfezione della saggezza. La coproduzione condizionata. Tradizioni secolarizzate. Pratiche secolarizzate di coltivazione della compassione. Il significato della secolarizzazione.

## Terza lezione: 28-30 Marzo 2025 

@ Istituto Lama Tzong Khapa di Pomaia

### Le pratiche contemplative nella relazione di aiuto (Prima Parte)

**Obiettivi** Comprendere compassionevolmente il funzionamento umano e sviluppare la qualità della compassione nella relazione di aiuto e nella propria vita quotidiana.

**Contenuti**

**Dott.ssa Elisa Ruggeri**  *Sviluppare disponibilità e accettazione -- risorse per il clinico nella relazione d’aiuto e per il paziente* 1. Perché disponibilità e accettazione? 2. L’evitamento esperienziale. 3. Creare un’apertura alla disponibilità. 4. L’atteggiamento terapeutico, la cura di se’ e degli altri.

**Dott. Massimo Gusmano** *La compassione e le pratiche che aiutano a svilupparla dal punto di vista del buddhismo* 1. cosa intende il buddhismo per compassione? 2. la relazione con gli altri Brahma Vihara e la coltivazione della parte etica. 3. pratiche di meditazione per sviluppare la compassione e aprire il cuore. 4. portare la compassione nella relazione di cura.

**Dott. Giovanni Zucchi** *Principi di Compassion-Focused Therapy (CFT) e Acceptance and Commitment Therapy (ACT)*  1. Il modello della Compassion Focused Therapy. 2. Mente evoluta e comprensione compassionevole: uno sguardo saggio sull’essere umano, sui suoi limiti e sulla sua evoluzione. 3. Le tecniche di self compassion e la maturazione delle abilità di cura di sé e degli altri. 4. Gli ostacoli alla compassione: evitamento, giudizio e inflessibilità psicologica. 5. Il lavoro personale tra disidentificazione e compassione. 

## Quarta lezione: 11 Aprile 2025 

@ UNIFI

### Le pratiche contemplative nella relazione di aiuto (Parte Seconda)

**Obiettivi** Consentire l’applicazione del modello fondato sulle qualità relazionali nella ristrutturazione cognitivo – emozionale e nel lavoro con gli stati emotivi.

**Contenuti** 

**Dott. Silverio Zucchi** *L’Acceptance and Commitment Therapy (ACT) e la ricerca di senso*  1. L’ubiquità della sofferenza umana, le sue attuali manifestazioni e la ricerca di senso. 2. Il modello della flessibilità psicologica nell’ACT. 3. La postura del clinico e il lavoro sui valori. 4. Qualità di una conversazione efficace sui valori. 5. Esercitazioni sui valori.

**Dott. Giovanni Zucchi** *Le qualità costruttive che maturano attraverso la meditazione* 1. L’attitudine meditativa nella relazione. 2. Le condizioni facilitanti la costruzione di una relazione efficace. 3. L’ascolto profondo e il rispecchiamento.

## Quinta lezione: 15 Aprile 2025 

@ UNIFI

### Il processo emotivo e la correlazione mente/corpo

- Dott. Massimo Gusmano 
- Dott.ssa Elisa Ruggeri

**Obiettivi** Un'introduzione alla psicologia delle emozioni e al suo ruolo all’interno della comprensione di sé e dell’altro. Approfondire il ruolo e la funzione delle pratiche contemplative nella relazione con le emozioni.

**Contenuti** 

**Dott.ssa Elisa Ruggeri** *Emozione e cognizione nella psicologia contemporanea* Il processo fondamentale del decentraggio o disidentificazione nella gestione emozionale.

**Dott. Massimo Gusmano** *Le emozione secondo la psicologia buddhista ed il ruolo delle pratiche di meditazione* La reattività emotiva e la trasformazione delle emozioni.  Lo sviluppo degli stati emotivi costruttivi: benevolenza, compassione, gioia altruistica ed equanimità.
